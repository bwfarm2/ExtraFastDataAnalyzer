```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = 'data.csv'
data = pd.read_csv(file_path, header=None)

# Extract time and voltage columns
time_values = data.iloc[:, 0].values
voltage_values = data.iloc[:, 1].values

# Plot raw data
plt.figure(figsize=(10, 6))
plt.plot(time_values, voltage_values)
plt.title('Raw Data')
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.grid(True)
plt.show()

# Calculate the sampling frequency
sampling_frequency = 1 / (time_values[1] - time_values[0])

# Perform FFT
fft_result = np.fft.fft(voltage_values)
frequencies = np.fft.fftfreq(len(fft_result)) * sampling_frequency

# Find the dominant frequency
dominant_frequency_index = np.argmax(np.abs(fft_result))
dominant_frequency = frequencies[dominant_frequency_index]

# Plot FFT result
plt.figure(figsize=(10, 6))
plt.plot(frequencies, np.abs(fft_result))
plt.title('FFT Result')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

print("Dominant Frequency:", dominant_frequency, "Hz")
```
